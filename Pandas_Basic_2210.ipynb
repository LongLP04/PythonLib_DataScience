## Data manipulation : thao tác với dữ liệu

### Apply()

```python
# Có những dữ liệu số ( như tiền ) sẽ ở dạng String, gây ảnh hưởng tới việc thu nhập dữ liệu 
df.item_price.dtype
# dtype('O') là String
```

```python
df.item_price = df.item_price.apply(lambda x : float(x.replace('$', '')))
# lambda : hàm dùng để nhận từng giá trị của hàng trên cột được chọn
# .replace(n,m) dùng để thay n thành m
# float() dùng để chuyển String sau khi replace thành float
```

```python
df.head()
df.item_price.dtype
```

### Tạo cột mới

```python
# Structure : df["new_name_col"] = df["name_col1"] * df["name_col2"]
df["total_price"] = df["quantity"] * df["item_price"]
```

```python
df.head()
```

### EX1 : How much was the revenue for the period in the dataset?


### Sum()

```python
# dùng Sum()
x = df["total_price"].sum()
print("x = " ,x)
# hoặc có thể dùng lob
y = df.loc[:, "total_price"].sum()
print("y = ",y)
# hoặc dùng iloc ( không nên vì sau này nếu thêm cột thì vị trí cột -1 sẽ khác ) nên dùng loc ( lable location )
z = df.iloc[:, -1].sum()
print("z = ",z)
```

### EX2 : What was the most-ordered item?


### Group By()

```python
c = df.groupby("item_name")["quantity"].sum()
c.sort_values(ascending = False).head()
# dùng groupby("name_col") để nhóm các row có cùng tên lại
# ["name_col2"].sum() để hiển thị col cần xét ( ở đây là số lượng )
# sort_values(ascending = False) để sắp xếp giảm dần, mặc định là tăng dần
```

### value_counts() tương tự groupby()

```python
df.item_name.value_counts().head()
# Structure : df.name_col.value_counts()
# Value_counts() tự động sắp DESC, không cần sort
```

```python
# Ngoài ra có thể tính số item_name khác nhau trong file bằng 2 cách
# Cách 1 : value_counts().count()
x = df.item_name.value_counts().count()
print("Số món tính bằng cách 1 : ", x)
# Cách 2 : dùng nunique()
y = df.item_name.nunique()
print("Số món tính bằng cách 2 : ", y)
```
