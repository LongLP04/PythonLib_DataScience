## Project : Data Visualazation - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("vietnam_provinces.csv")
cities['area_total_km2'].sort_values(ascending=False)
```

```python
# LẤY KINH ĐỘ (LONGD) VÀ VĨ ĐỘ (LATD) ĐỂ THỂ HIỆN RA BIỂU ĐỒ
lat, lon = cities['latd'], cities['longd']
lat.head()
lon.head()
```

```python
# thể hiện kinh và vĩ độ trong biểu đồ bằng python api

population, area = cities['population_total'], cities['area_total_km2']
plt.style.use('seaborn-v0_8-darkgrid')

plt.figure(figsize=(6,10));
plot1 = plt.scatter(lon, lat, 
            c = np.log10(population), cmap = "viridis",
            s = area/10, linewidths=0, alpha= 0.8);
plt.axis('equal');       
plt.xlabel("Kinh Độ")
plt.ylabel("Vĩ Độ")
plt.title("Bản đồ Việt Nam");
clb = plt.colorbar(plot1);
clb.set_label('Mức độ phân bổ dân cư', fontsize=12, labelpad=10)

# tạo thêm chú thích kích cỡ từng thành phố
area_arange = [500,1000,3000,8000]
for area in area_arange:
    plt.scatter([],[], s = area/20, label = str(area) + 'km', c = 'k', alpha = 0.3)
plt.legend(labelspacing =1, title = "Diện tích");
```
