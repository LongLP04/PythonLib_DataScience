## TỰ HỌC MATPLOTLIB

### Matplotlib : Dùng để trực quan hóa dữ liệu ở dạng tĩnh, động hay tương tác được trực tiếp ở dạng đồ thị

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Trang trí đồ thị cho phù hợp với từng loại
# Kiểm tra những style có sẵn trong matplotlib
plt.style.available

# Dùng style đó trong tất cả đồ thị dưới
plt.style.use("dark_background")


```

```python
# Tạo 1 đồ thị trống bằng .plot()
plt.plot();
```

```python
# Flot 1 đồ thị đơn giản
x = [1,2,3,4,5]
y = [8,12,3,5,6]

# Lưu ý : có thể thay đổi màu sắc của đường thẳng trong đồ thị bằng cách thêm tham số color
plt.plot(x,y, color= "green");
```

## 2 cách để vẽ đồ thị với Maplotlib

### Pyplot API & Ojbect-Oriented API
* Pyplot API : nhanh, dùng như ở trên
* OO API : dùng để vẽ các đồ thị nâng cao, phức tạp 

### PYPLOT API



```python
x = np.linspace(0,10,1000)


# Tạo đồ thị gồm 2 line sin và cos x
# Color thay đổi màu sắc line
#linestyle thay đổi style line
# label dùng để gắn nhãn chú thích cho line
plt.plot(x, np.sin(x), color = "Red", linestyle = "dashed", label="Sin(X)");
plt.plot(x, np.cos(x), color = "Blue", label = "Cos(X)");

# Thêm nhãn của 2 trục
plt.xlabel("X")
plt.ylabel("F(X)")

# Thêm title cho đồ thị
plt.title("Sin(X) & Cos(X)")

# Để hiện các chú thích của line phải dùng .legend()
plt.legend()

# Giới hạn range của đồ thị theo 2 cách
# Cách 1 :
#plt.xlim([0,5])
#plt.ylim([-0.75, 0.75])

# Cách 2 : 
#plt.axis([0,4,-0.75,0.75])

# Dùng axis("tight") để fix đồ thị chắc vào trong khung
# Dùng axis("equal") để fix cho trục tung hoành bằng nhau
#plt.axis("tight")
#plt.axis("equal")
```

##  OO API : dùng để vẽ đồ thị nâng cao

```python
# B1 : Chuẩn bị dữ liệu
x = [1,3,5,8]
y = [5,9,23,28]


# B2 : Chuẩn bị plot
fig, ax = plt.subplots(figsize =(5,5))
""" Giải thích 
Thay vì cho toàn cục bộ plt. như cách 1, ở đây có 2 thuộc tính là figure và axes
Có thể hiểu figure là 1 khoảng giấy trắng để chứa được các khung vẽ
Còn Axes là 1 khung vẽ trong fig đó, thao tác thêm sửa,... đều dùng trong ax hết
1 fig có thể có nhiều khung
VÌ VẬY NẾU CẦN VẼ NHIỀU ĐỒ THỊ TRONG CÙNG 1 KHUNG THÌ PHẢI DÙNG CÁC OO API NÀY
"""

# B3 : Plot dữ liệu 
ax.plot(x,y, color = "green");

# Set các thuộc tính trong plot, dùng ax.set()
ax.set(title = " Simple plot",
        xlabel = "X",
        ylabel = "Y")


"""
Cách này tương tự cách 1 nhưng nâng cao hơn vì có thể thao tác được trực tiếp từng biểu đồ thay vì chung chung như cách 1
"""
```

## Most common type of Matplotlib plots
* line
* scatter
* bar
* hist
* subplots()

### Line : Biểu đồ nối line thể hiện mối quan hệ giữa các điểm
* Thời gian (time series)
* Đường cong hàm số
* Dữ liệu liên tục

```python
# B1 Create an array
x = np.linspace(0,10,100)


# B2 : Setup plot
fig, ax = plt.subplots()


# B3 : plots the data x
ax.plot(x, x**3);
```

### Scatter : Biểu đồ phân tán không nối line với nhau
* Phân bố dữ liệu
* So sánh hai biến
* Phát hiện outlier, clusters

```python
# Đầu tiên thử vẽ bằng Pyplot trước
plt.scatter(x, x**2);

# kết quả trả về sẽ là những chấm tách rời thể hiện giá trị của từng điểm
```

```python
# OO API

# B1 : Chuẩn bị số liệu
rng = np.random.RandomState(100)

x = rng.randn(100)
y = rng.randn(100)

# trang trí plot
# tăng kích thước các hạt, mỗi hạt sẽ có 1 kích thước khác nhau
size = 200*rng.randn(100)
# thay đổi màu sắc các hạt, mỗi hạt sẽ có 1 màu sắc khác nhau
color = rng.randn(100)

# B2 setup plot
fig, ax = plt.subplots()

# B3 plot data
plot1 = ax.scatter(x,y, s = size, c = color, cmap = "cool", alpha = 0.7);
# cmap : color map : tương tự như tông màu gồm 'viridis', 'inferno', 'cool', 'hot', 'jet'
# alpha : độ trong suốt ( từ 0 - > 1 )

# Add thêm colorbar để xem với màu sắc này thì tương ứng với mức độ bao nhiêu
# thao tác trên fig
# lưu ý phải đặt cho đồ thị 1 tên để cung cấp tham số cho colorbar làm việc
fig.colorbar(plot1)
```
